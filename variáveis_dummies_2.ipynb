O BANCO QUE SOLICITOU O MODELO DE LIMITE DE CREDITO INCLUIU NOVAS VARIÁVEIS NO DATASET COMO ESCOLARIDADE, GENERO E ESTADO DO CLIENTE. FAÇA AS TRANSFORMAÇÕES NECESSÁRIAS NESSAS VARIÁVEIS DUMMIES E TIRE O MODELO OLS ATÉ QUE O MODELO FIQUE AJUSTADO.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
limite = pd.read_excel("/content/C├│pia de Limite_Credito_Dummy.xlsx")

limite.head()
```

```python
# TRANSFORMAR DADO CATEGÓRICO EM DUMMY:

limite = pd.get_dummies(data=limite, columns = ["Escolaridade", "Gênero", "Região"], prefix=["Escolaridade", "Gênero", "Região"])

limite.head()
```

```python
limite.corr()
```

```python
# VERIFICAR OLS:

import statsmodels.formula.api as smf
```

```python
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Gênero_RJ+Gênero_SP+Região_F+Região_M"
```

```python
model = smf.ols(formula=function, data=limite).fit()

print(model.summary())
```

```python
# COMEÇAR A EXCLUIR VARIÁVEL CFME O VALOR DO P-VALUE:

function = "LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Gênero_RJ+Gênero_SP+Região_F+Região_M"

model = smf.ols(formula=function, data=limite).fit()

print(model.summary())
```
